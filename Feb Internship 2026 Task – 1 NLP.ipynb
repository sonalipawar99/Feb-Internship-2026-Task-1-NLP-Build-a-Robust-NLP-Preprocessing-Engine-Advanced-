### Task 2:Build Advanced Preprocessing Function
### Implement the following function:def preprocess_text(text): pass
### Required Features:Your function must include:Remove numbers
### Example: "I have 2 dogs" → "I have dogs"Remove extra spaces
### Example: "This is good" → "This is good"Handle repeated characters
### Example: "soooo goooood!!!" → "so good"Convert text to lowercase
### Example: "WOW!!! This is GREAT!!!" → "wow this is great Remove very short tokens Remove words with length ≤ 2
### Exception: keep meaningful words like "no", "not"
### Remove URLs and email-like patterns
### Example: "Visit http://example.com now" → "visit now"

```python
import re

def preprocess_text(text):
    # Handle empty or invalid input
    if not text or not isinstance(text, str):
        return [], ""

    # 1. Convert to lowercase
    text = text.lower()

    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # 3. Remove email patterns
    text = re.sub(r'\S+@\S+', '', text)

    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 5. Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 6. Remove special characters & emojis
    text = re.sub(r'[^a-z\s]', '', text)

    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 8. Tokenization
    tokens = text.split()

    # 9. Remove short words (≤2) except "no", "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    # 10. Join back to sentence
    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence
print(preprocess_text("I absolutely looooved this product "))
```

```python
test_data = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product ",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for text in test_data:
    tokens, clean = preprocess_text(text)
    print("Original:", text)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    print("-"*50)
```

### Task 4: Token Analytics
### For each processed sentence, compute:Total number of tokens,Number of unique tokens,Average token length,Analysis Questions,Which sentence had the most noise?,Which sentence retained the most meaningful tokens after cleaning?


```python
for text in test_data:
    tokens, clean = preprocess_text(text)

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(w) for w in tokens)/total if total else 0

    print("Sentence:", text)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg_len, 2))
    print("-"*50)
```

### Task 5: Frequency Analysis
### Combine all tokens and use:from collections import Counter
### Identify:Top 10 most frequent words,Top 5 least frequent words

```python
from collections import Counter

all_tokens = []

for text in test_data:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 most frequent:", counter.most_common(10))
print("Top 5 least frequent:", counter.most_common()[-5:])
```

### Task 6: Build Full Pipeline
### def full_pipeline(text_list): pass
### Expected Output:{ "tokens": [...], "clean_sentences": [...]}

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        
        # collect all tokens
        all_tokens.extend(tokens)
        
        # collect cleaned sentences
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
   
test_data = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product ",
    "Worst service ever... 0/10"
]

result = full_pipeline(test_data)

print("Tokens:", result["tokens"])
print("Clean Sentences:", result["clean_sentences"])
```
